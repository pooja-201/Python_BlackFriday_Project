# MultiColumn Analysis we use SEABORN library

```python
import pandas as pd
import seaborn as sns
df=pd.read_csv(r"C:\lenovo data\Study\Black Friday\BlackFriday.csv")

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
# first we take age & gender column
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.countplot(x ='Age' ,hue ='Gender' , data=df)
```

```python
sns.countplot(x ='Gender' ,hue ='Age' , data=df)
```

# So we conclude that Males are have more purchasing power that Females irrespective of age

```python
sns.countplot(x = df['City_Category'])
```

```python
sns.countplot(x = 'City_Category',hue='Marital_Status',data=df)
```

```python
# Count of Unmarried people are more that married in each city
```

```python
# count of 26-35 age group people are more than other age groups
# we can reverse the axis like x = 'City_Category',hue='Age'
sns.countplot(x = 'Age',hue='City_Category',data=df)
```

```python
df.groupby('City_Category').sum()['Purchase'].plot(kind = 'pie', figsize = (6,6) ,autopct = "%.1f")
```
